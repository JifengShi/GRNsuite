# Burst detection for sugar responses

```python
import grnsuite.burst_detection as burst_detection
import os
from pathlib import Path

import matplotlib
matplotlib.use("TkAgg")  # Ensures the correct interactive backend is used

# Set input directory
input_dir = 'results/20231103-M05-sucr-100-Gal-A2-01'

# Create output directory
output_dir = os.path.join(input_dir, 'burst_analysis')

# Find the spike file and signal file 
spike_file = os.path.join(input_dir, 'detected_spikes.csv')
signal_file = os.path.join(input_dir, 'interactive_processed.csv')

# Process the file
results = burst_detection.process_spike_file(
    spike_file,  # Use the full path to the CSV file
    column_name='spike_times', 
    output_dir=output_dir
)

# Visualize the results
burst_detection.plot_burst_classification(
    signal_file=signal_file,
    burst_results=results,
    output_dir=output_dir,
    time_range=None,  # Plot the full signal
    figsize=(15, 7),
    save_fig=True,
    dpi=300
)
```
